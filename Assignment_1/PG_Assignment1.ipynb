**Ques 8:**

Find area of a triangle when lengths of all three sides are given (Heron's formula).

**Explanation**:

Heron's Formula is used to calculate the area of a triangle when the lengths of all three sides are known. Unlike the standard area formula, it does not require the height of the triangle.

**Formula:**

Area = √(s × (s − a) × (s − b) × (s − c))

where,

s = (a + b + c) / 2
- **a, b, c:** Lengths of the three sides of the triangle.
- **s:** Semi-perimeter, which is half of the triangle's perimeter.
- First, calculate the semi-perimeter (`s`).
- Then, substitute the values into Heron's formula to find the area.
- The square root of the result gives the area of the triangle.

```python
import math # importing built-in library

a=int(input("Enter a:"))   # a=10
b=int(input("Enter b:"))   #b=5
c=int(input("Enter c:"))   #c=7
s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("Area of triangle:", area)
```

**Ques 5:**

Check whether a quadratic equation has real or imaginary roots and display the roots.

Equation format: ax^2 + bx + c = 0

**Explanation:**


A quadratic equation is an equation of the form **ax² + bx + c = 0**, where **a ≠ 0**. The roots of the equation are calculated using the quadratic formula.
- **a, b, c:** Coefficients of the quadratic equation entered by the user.
- **Discriminant (D):** Calculated as **D = b² − 4ac**.
- If **D > 0**, the equation has **two distinct real roots**.
- If **D = 0**, the equation has **two equal real roots**.
- If **D < 0**, the equation has **two imaginary (complex) roots**.
- The program uses the `cmath` module to calculate both real and complex roots.

```python
import cmath

# Take input from the user
a = float(input("Enter coefficient a: ")) #4.5
b = float(input("Enter coefficient b: ")) #3.1
c = float(input("Enter coefficient c: ")) #2.3

# Calculate discriminant
d = b ** 2 - 4 * a * c

# Calculate roots
root1 = (-b + cmath.sqrt(d)) / (2 * a)
root2 = (-b - cmath.sqrt(d)) / (2 * a)

# Check the nature of roots
if d > 0:
    print("The roots are real and distinct.")
elif d == 0:
    print("The roots are real and equal.")
else:
    print("The roots are imaginary.")

# Display the roots
print("Root 1 =", root1)
print("Root 2 =", root2)
```

**Ques 8:**

Print the grade sheet of a student from marks of five subjects.

Percentage = (total marks / 500) * 100
CGPA = Percentage / 10
Grade is assigned using the provided CGPA ranges.

**Explanation:**

- **Total Marks** are calculated by adding the marks of all subjects.
- **Percentage** is calculated using the formula:

  Percentage = (Total Marks / 500) × 100

- **CGPA** is calculated as:

  CGPA = Percentage / 10

- The grade is assigned based on the calculated CGPA using predefined grade ranges.
- Finally, the program displays the student's grade sheet.

```python
# Function to calculate grade
def get_grade(cgpa):
    if 0.0 <= cgpa <= 3.4:
        return "F"
    elif 3.5 <= cgpa <= 5.0:
        return "C+"
    elif 5.1 <= cgpa <= 6.0:
        return "B"
    elif 6.1 <= cgpa <= 7.0:
        return "B+"
    elif 7.1 <= cgpa <= 8.0:
        return "A"
    elif 8.1 <= cgpa <= 9.0:
        return "A+"
    elif 9.1 <= cgpa <= 10.0:
        return "O (Outstanding)"
    else:
        return "Invalid CGPA"

# Take input from the user
name = input("Enter name: ")
roll = input("Enter roll number: ")

m1 = float(input("Enter marks of Subject 1: "))
m2 = float(input("Enter marks of Subject 2: "))
m3 = float(input("Enter marks of Subject 3: "))
m4 = float(input("Enter marks of Subject 4: "))
m5 = float(input("Enter marks of Subject 5: "))

# Calculate total, percentage and CGPA
total = m1 + m2 + m3 + m4 + m5
percentage = (total / 500) * 100
cgpa = round(percentage / 10, 1)
grade = get_grade(cgpa)

# Display result
print("\n----- Grade Sheet -----")
print("Name:", name)
print("Roll Number:", roll)
print("Total Marks:", total)
print("Percentage:", round(percentage, 2), "%")
print("CGPA:", cgpa)
print("Grade:", grade)
```

Check whether a given number is palindrome or not.

**Algorithm:**

1. Start the program.
2. Take a number `num` as input from the user.
3. Store the original number in another variable.
4. Reverse the number using a loop.
5. Compare the reversed number with the original number.
6. If both numbers are equal, the number is a palindrome.
7. Otherwise, the number is not a palindrome.
8. Stop the program.



```python
num = int(input("Enter a number: ")) #num=343
 
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print(original, "is a palindrome number")
else:
    print(original, "is not a palindrome number")
```

- A function `find_max_min()` is created to perform the calculation.
- The first element of the sequence is initially considered as both the maximum and minimum value.
- A loop checks each remaining value in the sequence.
- If a value is greater than the current maximum, it becomes the new maximum.
- If a value is smaller than the current minimum, it becomes the new minimum.
- The function returns both maximum and minimum values.
- The returned values are stored in separate variables and displayed.

```python
def find_limits(data):
    high = data[0]
    low = data[0]

    for item in data[1:]:
        if item > high:
            high = item

        if item < low:
            low = item

    return high, low


numbers = [24, 8, 56, 13, 91]

max_value, min_value = find_limits(numbers)

print("Maximum:", max_value)
print("Minimum:", min_value)
```

**Ques 5:**

Create multiple suitable exceptions for a file handling program.

**Explanation:**

- A custom exception is a user-defined exception created by inheriting from the `Exception` class.
- `EmptyFileError` is raised when the file contains no data.
- `InvalidDataError` is raised when a record has an incorrect format or invalid values.
- The program also handles `FileNotFoundError` if the file does not exist.
- Exception handling helps the program display meaningful error messages instead of terminating abruptly.


```python
class EmptyFileError(Exception):
    pass


class InvalidDataError(Exception):
    pass


def read_file(file_name):
    try:
        file = open(file_name, "r")

        lines = file.readlines()
        file.close()

    except FileNotFoundError:
        raise FileNotFoundError("File not found.")

    if len(lines) == 0:
        raise EmptyFileError("File is empty.")

    details = {}

    for line in lines:
        line = line.strip()

        parts = line.split(",")

        if len(parts) != 2:
            raise InvalidDataError("Invalid record: " + line)

        student = parts[0].strip()

        try:
            marks = float(parts[1].strip())

        except ValueError:
            raise InvalidDataError("Invalid marks: " + line)

        details[student] = marks

    return details


with open("record.txt", "w") as file:
    file.write("Aman,82\n")
    file.write("Neha,91\n")


try:
    result = read_file("record.txt")

    print("Student Details:")
    for name, marks in result.items():
        print(name, ":", marks)

except (FileNotFoundError, EmptyFileError, InvalidDataError) as error:
    print(error)
```

```python
class Student:

    def __init__(self, name, sap_id, physics, chemistry, maths):
        self.name = name
        self.sap_id = sap_id
        self.physics = physics
        self.chemistry = chemistry
        self.maths = maths

    def display(self):
        print("Name:", self.name)
        print("SAP ID:", self.sap_id)
        print("Physics:", self.physics)
        print("Chemistry:", self.chemistry)
        print("Maths:", self.maths)

    def marks_percentage(self):
        total = self.physics + self.chemistry + self.maths
        return total / 3

    def result(self):
        if self.physics > 40 and self.chemistry > 40 and self.maths > 40:
            return "Pass"
        else:
            return "Fail"


def class_average(student_list):
    total = 0

    for student in student_list:
        total = total + student.marks_percentage()

    return total / len(student_list)


s1 = Student("Aman", "5101", 82, 75, 91)
s2 = Student("Neha", "5102", 65, 88, 72)
s3 = Student("Rahul", "5103", 39, 70, 55)

students = [s1, s2, s3]

for student in students:
    student.display()
    print("Percentage:", round(student.marks_percentage(), 2))
    print("Result:", student.result())
    print()

print("Class Average:", round(class_average(students), 2))
```

**Ques 3:**

Create programs to implement different types of inheritance.

**Explanation:**

Inheritance allows one class to acquire the properties and methods of another class. It helps in code reusability and reduces duplication.

### 1. Single Inheritance
- One child class inherits from one parent class.

### 2. Multilevel Inheritance
- A class inherits from another class, and a third class inherits from it.

### 3. Multiple Inheritance
- One child class inherits from two or more parent classes.



     


```python
# Single Inheritance
class Vehicle:
    def start(self):
        print("Vehicle is starting...")


class Car(Vehicle):
    def drive(self):
        print("Car is moving.")


# Multilevel Inheritance
class Device:
    def power_on(self):
        print("Device is switched on.")


class Laptop(Device):
    def open_screen(self):
        print("Laptop screen opened.")


class GamingLaptop(Laptop):
    def play_game(self):
        print("Playing games...")


# Multiple Inheritance
class Singer:
    def sing(self):
        print("Singing a song.")


class Dancer:
    def dance(self):
        print("Performing dance.")


class Performer(Singer, Dancer):
    pass


print("Single Inheritance:")
c = Car()
c.start()
c.drive()

print("\nMultilevel Inheritance:")
g = GamingLaptop()
g.power_on()
g.open_screen()
g.play_game()

print("\nMultiple Inheritance:")
p = Performer()
p.sing()
p.dance()
```

Create a class for operator overloading which adds two Point objects where each point has x and y values.
Example:
P1(x=10, y=20)
P2(x=12, y=15)
P3 = P1 + P2 gives P3(x=22, y=35)

**Explanation:**

Operator overloading allows us to define how operators such as `+`, `-`, `*`, etc., work with objects of a user-defined class.
- A class is created to represent a point with x and y coordinates.
- The special method `__add__()` is used to overload the `+` operator.
- When two objects are added, their corresponding x and y values are added separately.
- A new object is returned containing the resulting coordinates.

```python
class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def display(self):
        print("x =", self.x, ", y =", self.y)


point1 = Point(8, 14)
point2 = Point(16, 11)

point3 = point1 + point2

print("First Point:")
point1.display()

print("\nSecond Point:")
point2.display()

print("\nResult after Addition:")
point3.display()
```
